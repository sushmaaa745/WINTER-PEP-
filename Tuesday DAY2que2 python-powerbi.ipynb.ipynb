```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
food = pd.read_csv("/content/indian_food.csv")
food.head()
```

```python
diet_count = food['diet'].value_counts()
plt.figure()
sns.barplot(x=diet_count.index, y=diet_count.values)
plt.title("Vegetarian vs Non-Vegetarian Dishes")
plt.xlabel("Diet Type")
plt.ylabel("Count")
plt.show()
```

```python
unique_dishes = food['name'].nunique()
print("Count of unique dishes:", unique_dishes)
```

```python
unique_flavors = food['flavor_profile'].nunique()
print("Count of unique flavor profiles:", unique_flavors)
```
